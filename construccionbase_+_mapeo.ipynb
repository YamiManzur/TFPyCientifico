2018.

Advertencia: ciertas modificaciones en las latitudes se hicieron directamente al csv por lo que si corremos el programa con la base original, ciertos hogares quedarán fuera de ciudad de Santa Fe

```python
#Abrimos el archivo directamente como un dataframe
import pandas as pd
mapa=pd.read_csv("C:/Users/Usuario/Desktop/tesina/bases_excels/basemapas.csv", encoding = "ISO-8859-1", sep=";")
```

```python
#importamos biblioteca para georeferenciar las direcciones
import geocoder
```

```python
#construí una nueva variable para poner la dirección como la necesitaba el buscador
address=mapa["a1_1a_calle"]+", "+mapa["CIUDAD"]+", "+mapa["Departamento"]
```

```python
#ciclo for para obtener las coordenadas de las direcciones con open street maps
latitudlongitud=[]
for direccion in address:
    g=geocoder.osm(direccion)
    latitudlongitud.append(g.latlng)
```

```python
#ciclo for para obtener las coordenadas de ls direcciones con arcgis
latitudlongitud1=[]
for direccion in address:
    g=geocoder.arcgis(direccion)
    latitudlongitud1.append(g.latlng)
```
